```python
import numpy as np 
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('mat_apriori.csv')
```

```python
freq_its = apriori(df, min_support=0.25, use_colnames=True)
```

```python
freq_its
```

```python
rules=association_rules(freq_its, metric="confidence", min_threshold=0.60)
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
rules = rules.sort_values(['confidence'], ascending=False)
```
