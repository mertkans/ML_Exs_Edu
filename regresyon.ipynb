```python
X=df.drop(['G3','G3C'], axis=1)
```

```python
X.shape
```

```python
y=df[['G3']]
```

```python
from sklearn.linear_model import LinearRegression as lrn
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.3, random_state=312)
```

```python
lr=lrn()
regmodel=lr.fit(X_train,y_train)
y_predict=regmodel.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2sonuc=r2_score(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)**0.5
```
